# Preprocess Function

```python
def preprocess(text):
    doc = nlp(text)
    filtered_text = []

    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_text.append(token.lemma_)

    return ' '.join(filtered_text)
```

```python
df['Preprocessed_comments'] = df['Comment'].apply(preprocess)
```

## Encoding

```python
encoder = LabelEncoder()
df['Emotion Number'] = encoder.fit_transform(df['Emotion'])
```

```python
df
```

# Featrue Selection

```python
X = df['Preprocessed_comments']
y = df['Emotion Number']
```

# Split Data

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=57)
```

# Featrue Engineering

```python
v = TfidfVectorizer() # Factorization (convert into numeric values )
X_train_cv = v.fit_transform(X_train) # fit => (mapping of unique terms to indices), transform => Converts the text data in X_train into a sparse matrix (rows = documents, columns = terms)
X_test_cv  = v.transform(X_test)
```

```python
X_train_cv.toarray()
```

# ML Models

```python
models = {
    "LogisticRegression": LogisticRegression(),
    "SupportVectorMachine": SVC(),
    "DecisionTree": DecisionTreeClassifier(),
    "RandomForest": RandomForestClassifier(),
    "BaggingClassifier": BaggingClassifier(),
    "VotingClassifier": VotingClassifier(estimators=[
        ('lr', LogisticRegression()),
        ('svc', SVC(probability=True)),
        ('dt', DecisionTreeClassifier())],
        voting='soft'),
    "KNeighbors": KNeighborsClassifier(),
    "CatBoost": CatBoostClassifier(silent=True),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'),
    "LightGBM": LGBMClassifier()
}

results = []
for model_name, model in models.items():
    print(f"\nTraining and evaluating model: {model_name}")
    try:
        model.fit(X_train_cv, y_train)
        y_pred = model.predict(X_test_cv)
        acc = accuracy_score(y_test, y_pred)

        print("Classification Report:")
        print(classification_report(y_test, y_pred))
        print("Confusion Matrix:")
        print(confusion_matrix(y_test, y_pred))

        results.append({
            'Model': model_name,
            'Accuracy': acc * 100,
        })

    except Exception as e:
        print(f"Error training {model_name}: {e}")

results_df = pd.DataFrame(results)
display(results_df)
```

# Cross-validation


```python
from sklearn.model_selection import cross_val_score

# Perform cross-validation
results = []
for model_name, model in models.items():
    print(f"\nTraining and evaluating model: {model_name}")
    try:
        scores = cross_val_score(model, X_train_cv, y_train, cv=5, scoring='accuracy') # 5-fold cross-validation
        avg_accuracy = np.mean(scores) *100
        print(f"Cross-validation scores: {scores}")
        print(f"Average Accuracy: {avg_accuracy:.2f}%")
        results.append({
            'Model': model_name,
            'Accuracy': avg_accuracy,
        })
    except Exception as e:
        print(f"Error training {model_name}: {e}")

results_df = pd.DataFrame(results)
display(results_df)
```

```python
# chat wth chatbot

def chat_with_bot(user_input):
    # Preprocess user input
    preprocessed_input = preprocess(user_input)

    # Transform the input using the same vectorizer
    input_cv = v.transform([preprocessed_input])

    # Predict the emotion using the best model (e.g., CatBoost)
    predicted_emotion_number = models["CatBoost"].predict(input_cv)[0]

    # Convert the predicted emotion number back to the emotion label
    predicted_emotion = encoder.inverse_transform([predicted_emotion_number])[0]

    return predicted_emotion

# Example usage:
user_input = "I'm feeling really good today."
predicted_emotion = chat_with_bot(user_input)
print(f"Predicted emotion: {predicted_emotion}")
```
